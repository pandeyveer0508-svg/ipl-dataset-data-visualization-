## i Match won by the maximum margin of runs

```python
 
import matplotlib.pyplot as plt
import seaborn as sb
max_win=ipl_matchs.sort_values("win_by_runs",ascending=False).head(5)
y=max_win["win_by_runs"]
x=max_win["id"]
sb.barplot(x=x,y=y,color="yellow",edgecolor="black")
plt.title(" Match won by the maximum margin of runs",color="red")
plt.ylabel("RUNS",color="green")
plt.xlabel("Match_id",color="darkgreen")
plt.show()
```

## x> Does winning the toss mean winning the match? 

```python

toss_match_win=(ipl_matchs["toss_winner"] == ipl_matchs["winner"])
import matplotlib.pyplot as plt
import seaborn as sb
toss_match_win=toss_match_win.map({True:"Win",False:"Lose"})
sb.countplot(x=toss_match_win,hue=toss_match_win,edgecolor="black")
plt.xlabel("Result")
plt.title("Win Toss Win Match",color="black",fontsize=16,fontstyle="italic",fontweight="bold",loc="center",pad=20)
plt.show()
```

Insights:
1. Toss helps, but it does not guarantee a win.
2. Many teams lost even after winning the toss.
3. Performance matters more than luck.
4. Toss may give slight advantage.

## xv> Best venue for defending and chasing a total. 

```python
top_venue_chase=(ipl_matchs["win_by_runs"]>0).groupby(ipl_matchs["venue"]).sum().sort_values(ascending=False).head()
import matplotlib.pyplot as plt
import seaborn as sb
x=top_venue_chase.index
y=top_venue_chase.values
sb.barplot(x=y,y=x,palette="viridis",edgecolor="black",linewidth=2,hue=top_venue_chase)
plt.ylabel("Venue",fontsize=16,color="black",fontstyle="italic")
plt.xlabel("No Of Matches Win",color="black",fontsize=16,fontstyle="italic")
plt.title("Venue VS Chasing win",color="black",fontsize=20,fontstyle="italic",loc="left",pad=20,fontweight="bold")
plt.show()
```

```python
top_venue_defend=(ipl_matchs["win_by_wickets"]>0).groupby(ipl_matchs["venue"]).sum().sort_values(ascending=False).head()
import matplotlib.pyplot as plt
import seaborn as sb
x=top_venue_defend.values
y=top_venue_defend.index
sb.barplot(x=x,y=y,palette="viridis",hue=top_venue_defend,edgecolor="black",linewidth=2)
plt.ylabel("Venue",color="black",fontsize=16,fontstyle="italic")
plt.xlabel("No Of Matches Win",color="black",fontsize=16,fontstyle="italic")
plt.title("Venue VS Defending Win",color="black",fontweight="bold",fontsize=20,fontstyle="italic",loc="left")
```

Insights:
1. Wankhede Stadium and Feroz Shah Kotla are the best venues for chasing runs, as teams often win while batting second there.
2. Eden Gardens and M. Chinnaswamy Stadium are better for defending totals, where teams batting first have a higher chance of winning.

## xvi> Best defending and chasing teams. 


```python
best_defending_teams=(ipl_matchs["win_by_wickets"]!=0).groupby(ipl_matchs["winner"]).sum().sort_values(ascending=False).head(10)
import matplotlib.pyplot as plt 
import seaborn as sb
x=best_defending_teams.index
y=best_defending_teams.values
sb.barplot(x=y,y=x,hue=best_defending_teams,palette="pastel",edgecolor="black")
plt.ylabel("Winner_Teams",color="red",fontsize=16,fontstyle='italic')
plt.xlabel("No Of Win",color="red",fontsize=16,fontstyle="italic")
plt.title("Win By Defending Vs Teams",color="black",fontsize=20,fontstyle="italic",fontweight="bold",loc="left",pad=20)
plt.show()
```

```python
best_chasing_teams=(ipl_matchs["win_by_runs"]!=0).groupby(ipl_matchs["winner"]).sum().sort_values(ascending=False).head(10)
import matplotlib.pyplot as plt
import seaborn as sb
x=best_chasing_teams.values
y=best_chasing_teams.index
sb.barplot(x=x,y=y,hue=best_chasing_teams,palette="pastel",edgecolor="black")
plt.ylabel("Winning Teams",color="red",fontsize=16,fontstyle="italic")
plt.xlabel("No Of Win",color="red",fontsize=16,fontstyle="italic")
plt.title("Win_By_Chasing VS Teams",color="black",fontsize=20,fontstyle="italic",fontweight="bold",loc="left",pad=20)
plt.show()
```

Insights:
1. Kolkata Knight Riders and Mumbai Indians have been strong teams while chasing targets.
2. Chennai Super Kings and Mumbai Indians have also performed well in defending totals.
